```python
#Cleaning Data: Casting Datatypes and Handling Missing Values
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Data/survey_results_public.csv')
df
```

```python
#Removing missing values: NaN, None
```

```python
df.dropna().head(3) #drops the whole rows, not a good one
```

```python
df.dropna(axis=0,how='any') #any missing value remove the whole row
```

```python
df.dropna(axis=0, how='all')# only if all values ina  row are missing or none then remove

#axis=0 or axis=index both gives the same results
```

```python
df.dropna(axis=1, how='all')# only if all values ina  COLUMN are missing or none then remove
#axis=1 or axis='columns' same thing
```

```python
df.dropna(axis=1, how='any')# any value is missing in column whole column is removed
```

```python
df.replace('NA',np.nan).head(5) #replaces all na with nan values
df.replace('Missing',np.nan).head(5) #replaces all na with nan values
```

```python
df.isna() #check if it has any missing values
```

```python
df.isnull().values.any() # does the data set has missing values
df.isna().values.any() # does the data set has missing values
```

```python
df.fillna('Missing')
#For mathematical purposes, we can replace null or na to 0
df.fillna(0)
```

```python
df['Age'].astype(int) #none values cant be converted
```

```python
df['Age']=df['Age'].astype(float) #none values cant be converted
```

```python
df['Age'].mean()
```

```python
df.columns
```

```python
df['YearsCode'].unique() #Less than 1 year and more than 50 years, so cant cast data type
```

```python
df['YearsCode'].replace ('Less than 1 year',0,inplace=True)
df['YearsCode'].replace ('More than 50 years',51,inplace=True)
```

```python
df['YearsCode']=df['YearsCode'].astype('float')
```

```python
df['YearsCode'].mean()
```
